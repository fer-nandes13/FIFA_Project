```python
#checking for multicolinearity using the VIF criteria (VIF > 5 signals multicolinearity)
# load statmodels functions
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
# creating a function for checking for vif:
def compute_vif(considered_features):
    X = numerical[considered_features]
    # the calculation of variance inflation requires a constant
    X['intercept'] = 1
    # create dataframe to store vif values
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable']!='intercept']
    return vif
```

## Creating function for df cleaning:

```python
#creating the function to clean the dataframe:

def df_clean(df):
    data = df
    #lowering_case the name of columns
    cols = []
    for column in data.columns:
        cols.append(column.lower())
    data.columns = cols
    #dropping columns that we will not use after our EDA
    col_ex = ['ID', 'Name', 'Nationality', 'Club', 'Position', 'Team & Contract', 'Loan Date End', 'Contract', 'LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB', 'LCB', 'CB', 'RCB', 'RB','GK', 'height', 'gk kicking', 'gk reflexes', 'gk handling', 'gk diving', 'goalkeeping', 'gk positioning', 'balance', 'acceleration', 'weight', 'sprint speed', 'w/f', 'wage', 'release clause', 'total stats', 'agility']
    col_ex = [x.lower() for x in col_ex]
    #deleting the selected columns
    data2 =data.drop(col_ex, axis = 1)
    #replacing the space for '_'
    converter = lambda x: x.replace(' ', '_')
    col_low= data2.columns
    col_low = list(map(converter, col_low))
    data2.columns = col_low
    #working with 'joined': we want to extract the year only
    data2[['joined_1', 'joined_2']] = data2['joined'].str.split(",", expand=True)
    data2['joined_2'] = pd.to_numeric(data2['joined_2'])
    data2["joined"] = data2["joined_2"]
    data2 = data2.drop(["joined_1", "joined_2"], axis = 1)
    #working with the column "value":
    data2['value'] = data2['value'].apply(convert_value)
    #cleaning sm' and 'ir' that contains a character '★':
    data2['sm'] = data2['sm'].str.replace("★", "").astype(int)
    data2['ir'] = data2['ir'].str.replace("★", "").astype(int)
    # Apply the function to the 'Hits' column and create a new 'NumericValue' column
    data2['hits'] = data2['hits'].apply(convert_hits)
    #deleting the 41 NaNs in joined as these observations were also missing for other 8 columns:
    data2.dropna(subset=['joined'], inplace=True)
    #a/w and d/w: replacing the 18 NaN with the mode:
    data2['a/w'] = data2['a/w'].fillna('Medium')
    data2['d/w'] = data2['d/w'].fillna('Medium')
    #Cleaning composure: **(check if works)**
    data2['composure'] = data2['composure'].fillna(data2['composure'].mean())
    #Splitting the data into numerical and categorical
    numerical = data2.select_dtypes(include = np.number)
    categorical = data2.select_dtypes(include = object)
    #X-y split:
    y = numerical['ova']
    X_num = numerical.drop('ova', axis = 1)
    X_cat = categorical
    #scaling the numerical variables using minmaxscaler:
    x_standardized = MinMaxtransformer.transform(X_num)
    x_numerical_s = pd.DataFrame(x_standardized, columns=X_num.columns)
    #now working with the categorical variables:
    cols = encoder.get_feature_names_out(input_features=X_cat.columns)
    encoded = encoder.transform(X_cat).toarray()
    onehot_encoded = pd.DataFrame(encoded, columns=cols)
    #Concating DataFrames
    X = pd.concat([x_numerical_s, onehot_encoded], axis=1)
    X.columns = X.columns.astype(str)
    return X, y
    
```
